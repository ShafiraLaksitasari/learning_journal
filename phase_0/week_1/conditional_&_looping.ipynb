### Control Flos Statement


Continue

```python
kl = 0

while kl <= 10:
    kl = kl + 1
    if kl % 2 == 1:
        continue  #
    print(kl)
```

Break

```python
km = 0

while km <= 10:
    km = km + 1
    print(km)
    if km == 7:
        break
```
